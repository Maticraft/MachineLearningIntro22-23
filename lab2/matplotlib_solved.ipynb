### Task 4 (1.0 pt) 
Use `subplots` to create six plots in the same figure, using two row and three column configuration. Plot $\sin(x)$, $\cos(x)$ or any any custom function in these plots. The plots should be correctly labeled and have the appropriate limits for easy viewing.

```python
fig, ax = plt.subplots(2,3, figsize = (10,10))
x = np.arange(0.5, 1, 1e-4)
for i, ax in enumerate(fig.axes):
    ax.plot(x, np.power(x, -1.0/(i+1)), label = 'x^(-1/%d)' % (i+1))
    ax.set_xlim([0.5,1])
fig.legend()
```

### Task 6 (1.0 pt)

Using the following functions to generate lines, use `fill_between` to fill regions between these lines in different combinations, i.e. use separate fill between $y_1$ and $y_2$, between $y_2$ and $y_3$, and between $y_1$ and $y_3$. For each area use a different color and use a good parameter `alpha` (by trial and errors) to make all areas visible.

The lines are given by:

$y_1=x^2-5$

$y_2=e^x+2$

$y_3=cos(x)$


where $x \in [0,3]$.

```python
x = np.arange(0, 3, 1e-4)
y1 = np.power(x, 2) - 5
y2 = np.exp(x) + 2
y3 = np.cos(x)
plt.fill_between(x, y1, y2, color='blue', alpha = 0.1)
plt.fill_between(x, y2, y3, color='green', alpha = 0.3)
plt.fill_between(x, y1, y3, color='magenta', alpha = 0.1)
```

### Task 7 (1.5 pt)
The shape of a limaon can be defined parametrically as

$r = r_0 + \cos\theta$

$x  = r \cos\theta$

$y  = r \sin\theta$


When $r_0 = 1$, this curve is called a cardioid. Use this definition to plot the shape of a limaon for $r_0 = 0.8$, $r_0 = 1.0$, and $r_0 = 1.2$. Be sure to use enough points that the curve is closed and appears smooth (except for the cusp in the cardioid). Use a legend to identify which curve is which. Save the plot as a `.pdf` file.

```python
thetas = np.arange(-np.pi, np.pi, 1e-4)

for r0 in [0.8, 1.0, 1.2]:
    r = r0 + np.cos(thetas)
    x = r * np.cos(thetas)
    y = r * np.sin(thetas)
    color = np.random.random(3)
    plt.plot(x, y, label = f'r0={r0}', c = color)
    plt.fill_between(x, -y, y, alpha = 0.3, color = color)
plt.legend()
```
